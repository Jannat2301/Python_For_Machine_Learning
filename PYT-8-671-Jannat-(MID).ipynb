# Detect Outliers

```python
import pandas as pd

# Function to check outliers in numerical columns
def detect_outliers_iqr(df):
    outlier_info = {}

    # Select numerical columns
    numerical_columns = df.select_dtypes(include=['int64', 'float64']).columns

    for column in numerical_columns:
        # Calculate Q1 (25th percentile) and Q3 (75th percentile)
        Q1 = df[column].quantile(0.25)
        Q3 = df[column].quantile(0.75)
        IQR = Q3 - Q1

        # Calculate lower and upper bounds for outlier detection
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        # Find rows with outliers
        outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]

        # Store results
        outlier_info[column] = {
            "num_outliers": len(outliers),
            "outlier_indices": outliers.index.tolist(),
            "lower_bound": lower_bound,
            "upper_bound": upper_bound,
        }
        
    return outlier_info

# Detect outliers in the dataset
outliers = detect_outliers_iqr(df)

# Display outlier information
for col, info in outliers.items():
    print(f"Column: {col}")
    print(f"  Number of Outliers: {info['num_outliers']}")
    print(f"  Outlier Indices: {info['outlier_indices'][:5]}")  # Show only first 5 indices
    print(f"  Lower Bound: {info['lower_bound']}, Upper Bound: {info['upper_bound']}")
    print("-" * 50)
```

# Handling Outliers 

```python
import pandas as pd

# Function to replace outliers with the mean of the respective column
def replace_outliers_with_mean(df):
    for column in df.select_dtypes(include=['float64', 'int64']).columns:  # Only for numerical columns
        # Calculate Q1, Q3, and IQR for the column
        Q1 = df[column].quantile(0.25)
        Q3 = df[column].quantile(0.75)
        IQR = Q3 - Q1
        
        # Define the lower and upper bounds for outliers
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        
        # Calculate the mean of the column
        mean_value = df[column].mean()

        # Replace outliers with the mean value
        df[column] = df[column].apply(lambda x: mean_value if x < lower_bound or x > upper_bound else x)
    
    return df

# Assuming your dataframe is named df
df_cleaned = replace_outliers_with_mean(df)

# Display the cleaned dataframe
print(df_cleaned.head())
```

# Conducting statistical analysis, including measures of central tendency, dispersion, and correlation.

```python
import pandas as pd

# 1. Measures of Central Tendency

# Mean, Median, and Mode for numeric columns
central_tendency = df.describe().transpose()[['mean', '50%', 'std']]

# Mode for each column (not covered by describe())
mode_values = df.mode().iloc[0]  # Get the first mode for each column
central_tendency['mode'] = mode_values

# 2. Measures of Dispersion
# Select only numeric columns
numeric_columns = df.select_dtypes(include=['number']).columns

# Range and IQR (Interquartile Range) for numeric columns
dispersion = pd.DataFrame({
    'Range': df[numeric_columns].max() - df[numeric_columns].min(),
    'IQR': df[numeric_columns].quantile(0.75) - df[numeric_columns].quantile(0.25),
    'Variance': df[numeric_columns].var(),
    'Standard Deviation': df[numeric_columns].std()
})

# 3. Correlation Analysis (for numeric variables)
correlation_matrix = df[numeric_columns].corr()

# 4. Combine all statistics into one DataFrame for easier interpretation
statistical_analysis = central_tendency.join(dispersion)

# Display all the statistics
print("Statistical Analysis:")
print(statistical_analysis)

print("\nCorrelation Matrix:")
print(correlation_matrix)
```

# Perform advanced grouping and aggregation, exploring relationships between multiple variables.

```python
import pandas as pd

# 1. Grouping by 'Course' and calculating the mean of numeric columns
grouped_by_course = df.groupby('Course')[['CGPA', 'Stress_Level', 'Depression_Score', 'Anxiety_Score']].mean()

# 2. Grouping by 'Gender' and calculating the median of numeric columns
grouped_by_gender = df.groupby('Gender')[['CGPA', 'Stress_Level', 'Depression_Score', 'Anxiety_Score']].median()

# 3. Grouping by 'Residence_Type' and calculating the standard deviation of numeric columns
grouped_by_residence = df.groupby('Residence_Type')[['CGPA', 'Stress_Level', 'Depression_Score', 'Anxiety_Score']].std()

# 4. Grouping by multiple columns (e.g., 'Course' and 'Gender') and calculating the sum of numeric columns
grouped_by_course_gender = df.groupby(['Course', 'Gender'])[['CGPA', 'Stress_Level', 'Depression_Score', 'Anxiety_Score']].sum()

# 5. Aggregating using multiple functions (mean, std, and count) on numeric columns
aggregation = df.groupby('Course').agg({
    'CGPA': ['mean', 'std', 'count'],
    'Stress_Level': ['mean', 'std'],
    'Depression_Score': ['mean', 'std'],
    'Anxiety_Score': ['mean', 'std']
})

# 6. Exploring relationships between 'Stress_Level', 'Anxiety_Score', 'Depression_Score', and 'CGPA' by course
relationship_by_course = df.groupby('Course')[['Stress_Level', 'Anxiety_Score', 'Depression_Score', 'CGPA']].corr()

# 7. Example of value counts to understand the distribution of categorical variables
value_counts_course = df['Course'].value_counts()
value_counts_gender = df['Gender'].value_counts()

# Display the results
print("Grouped by Course (Mean values):")
print(grouped_by_course)

print("\nGrouped by Gender (Median values):")
print(grouped_by_gender)

print("\nGrouped by Residence Type (Standard Deviation):")
print(grouped_by_residence)

print("\nGrouped by Course and Gender (Sum of values):")
print(grouped_by_course_gender)

print("\nAggregation by Course (Mean, Std, Count):")
print(aggregation)

print("\nCorrelation by Course between Stress, Anxiety, Depression, and CGPA:")
print(relationship_by_course)

print("\nValue Counts for Course:")
print(value_counts_course)

print("\nValue Counts for Gender:")
print(value_counts_gender)
```

```python
# Spearman's rank correlation for Depression_Score and CGPA for each course
from scipy.stats import spearmanr
correlation_results = {}

for course in df['Course'].unique():
    course_data = df[df['Course'] == course]
    correlation, p_value = spearmanr(course_data['Depression_Score'], course_data['CGPA'])
    correlation_results[course] = {'correlation': correlation, 'p_value': p_value}

# Display results
for course, result in correlation_results.items():
    print(f"{course} - Spearman Correlation: {result['correlation']}, P-value: {result['p_value']}")
```
